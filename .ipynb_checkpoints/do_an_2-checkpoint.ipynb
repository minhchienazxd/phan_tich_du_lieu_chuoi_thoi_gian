```python
# Hàm để xác định và hiển thị ngoại lai
def show_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Hiển thị các giá trị ngoại lai
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    print(f'Ngoại lai ở cột {column}:')
    return outliers[[column]]

# Hiển thị ngoại lai cho các cột charges
outliers_Temperature = show_outliers(data, 'nhiet_do')
outliers_Humidity = show_outliers(data, 'do_am')
outliers_general_diffuse_flows = show_outliers(data, 'dong_buc_xa_chung')
outliers_diffuse_flows = show_outliers(data, 'dong_buc_xa_tx')


# In các giá trị ngoại lai
print(outliers_Temperature)
print(outliers_Humidity)
print(outliers_diffuse_flows)
print(outliers_general_diffuse_flows)
```

```python
# Chuyển đổi cột 'DateTime' thành kiểu datetime và đặt làm chỉ mục
data['DateTime'] = pd.to_datetime(data['DateTime'])
data = data.set_index('DateTime')

data.info()
```

- ta sẽ dữ đoạn mức tiêu thụ điện năng cho 1 tháng tiếp theo 
- vì vậy tập test của ta sẽ có 4320 bản ghi 

```python
# Chọn các cột đặc trưng (exclude DateTime)
features = ['nhiet_do', 'do_am', 'toc_do_gio', 'dong_buc_xa_chung', 'dong_buc_xa_tx']
X = data[features].values

# Giả sử bạn muốn dự đoán khu1
y = data['khu1'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
# Chuẩn hóa dữ liệu
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))
```

```python
# Chia dữ liệu thành train-test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.08241758241758242, shuffle=False)
```

```python
# Tạo dữ liệu dạng chuỗi thời gian
import numpy as np

def create_sequence(data, target, time_steps=1):
    X_seq, y_seq = [], []
    for i in range(len(data) - time_steps):
        X_seq.append(data[i:i+time_steps])  # Lấy time_steps thời điểm trước
        y_seq.append(target[i+time_steps])  # Lấy nhãn tại thời điểm tiếp theo
    return np.array(X_seq), np.array(y_seq)
```

```python
time_steps = 1008  # Sử dụng 144 thời điểm trước để dự đoán

# Tạo chuỗi cho Train
X_train_seq, y_train_seq = create_sequence(X_train, y_train, time_steps)
print("X_train_seq shape:", X_train_seq.shape)  # (samples, time_steps, features)
print("y_train_seq shape:", y_train_seq.shape)  # (samples,)

# Tạo chuỗi cho Test
X_test_seq, y_test_seq = create_sequence(X_test, y_test, time_steps)
print("X_test_seq shape:", X_test_seq.shape)  # (samples, time_steps, features)
print("y_test_seq shape:", y_test_seq.shape)  # (samples,)
```

- Timestep là số bước thời gian sử dụng làm đầu vào để dự đoán các giá trị tương lai.
- Nó phụ thuộc vào:
    + Tần suất ghi dữ liệu (mỗi 10 phút).
    + Phạm vi thời gian bạn muốn mô hình nhìn lại (historical window).

### mô hình LSTM

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
model = Sequential()
model.add(LSTM(units=64, return_sequences=True,kernel_regularizer=l2(0.001), input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
model.add(Dropout(0.5))# Thêm dropout để giảm overfitting

model.add(LSTM(units=64, return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.001)))

model.add(Dense(1))  # Lớp đầu ra

model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Huấn luyện mô hình
model.fit(X_train_seq, y_train_seq, epochs=100, batch_size=32, validation_data=(X_test_seq, y_test_seq), callbacks=[early_stopping])
```

```python
model.summary()
```

```python
# Dự đoán trên tập kiểm tra
y_pred = model.predict(X_test_seq)

# Chuyển đổi giá trị dự đoán về phạm vi ban đầu (ngược chuẩn hóa)
y_pred_inversed = scaler_y.inverse_transform(y_pred)
y_test_inversed = scaler_y.inverse_transform(y_test_seq)

datetime_series = data.index[-len(y_test_seq):]  # Lấy DateTime tương ứng với y_test

# Tạo DataFrame chứa các giá trị thực tế và dự đoán cùng với DateTime
result_df = pd.DataFrame({
    'DateTime': datetime_series,  # Chuỗi thời gian từ dữ liệu gốc
    'Actual': y_test_inversed.flatten(),  # Giá trị thực tế
    'Predicted': y_pred_inversed.flatten()  # Giá trị dự đoán
})

# Hiển thị kết quả
print(result_df.head())  # Hiển thị 10 bản ghi đầu tiên
```

```python
plt.figure(figsize=(14, 7))
plt.plot(result_df['DateTime'], result_df['Actual'], label='Actual', color='blue')
plt.plot(result_df['DateTime'], result_df['Predicted'], label='Predicted', color='red', linestyle='--')
plt.xlabel('DateTime')
plt.ylabel('Value')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.xticks(rotation=45)  # Xoay các nhãn thời gian nếu cần
plt.grid(True)
plt.show()
```

- So sánh giữa giá trị thực và dự đoán:

    + Đường màu xanh (Actual - giá trị thực) và đường màu đỏ (Predicted - giá trị dự đoán) có sự trùng khớp tương đối tốt, cho thấy mô hình dự đoán khá chính xác.
Các đỉnh và đáy của dữ liệu (chu kỳ biến động lớn nhất) được mô hình dự đoán theo đúng xu hướng.

- Tính chu kỳ:

    + Biểu đồ thể hiện rõ ràng chu kỳ lặp lại hằng ngày trong dữ liệu, với sự tăng giảm định kỳ.
    + Mô hình dự đoán bắt được tính chu kỳ này, nhưng có một số khác biệt nhỏ.
- Sai lệch tại các đỉnh và đáy:

    + Ở một số đỉnh (giá trị cực đại) và đáy (giá trị cực tiểu), đường dự đoán (đỏ) hơi lệch so với giá trị thực (xanh).
    + Điều này cho thấy mô hình chưa hoàn toàn nắm bắt được toàn bộ sự thay đổi biên độ trong dữ liệu, đặc biệt ở những điểm cực trị.
- Biên độ biến động:

    + Dữ liệu thực có xu hướng biến động mạnh mẽ hơn (biên độ lớn hơn) so với đường dự đoán, đặc biệt trong một số khoảng thời gian ngắn.
    + Điều này có thể chỉ ra rằng mô hình đang làm mịn dữ liệu quá mức hoặc chưa nắm bắt đủ các yếu tố gây ra biến động lớn.
- Tổng thể hiệu suất:

    + Dự đoán nhìn chung theo xu hướng tốt, nhưng các khác biệt nhỏ ở chi tiết (đỉnh, đáy và biên độ) có thể cần cải thiện.
