```python
# Import các thư viện cần thiết
import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
# Thiết lập các tham số cho hàm y = mx + b + noise
m = 2  # Hệ số góc
b = 3  # Hệ số chặn
x = np.linspace(0, 50, 100)  # Tạo dữ liệu x từ 0 đến 50 với 100 giá trị
```

```python
# Đặt seed để đảm bảo tái tạo được noise
np.random.seed(101)
noise = np.random.normal(loc=0.0, scale=4.0, size=len(x))
```

```python
y = 2*x + b + noise
```

```python
plt.plot(x,y,'*')
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Tạo mô hình Sequential
model = Sequential()

# Thêm các lớp Dense vào mô hình
# Lớp đầu tiên có 4 nơ-ron, đầu vào có 1 feature (input_dim=1), và activation function là 'relu'
model.add(Dense(4, input_dim=1, activation='relu'))

# Lớp thứ hai có 4 nơ-ron và activation function là 'relu'
model.add(Dense(4, activation='relu'))

# Lớp cuối cùng có 1 nơ-ron (đầu ra), sử dụng activation function là 'linear'
model.add(Dense(1, activation='linear'))

# Compile mô hình
model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

```python
# Huấn luyện mô hình
model.fit(x, y, epochs=500, verbose=1)
```

```python
# Truy xuất lịch sử mất mát (loss) trong quá trình huấn luyện
loss = model.history.history['loss']

# Tạo danh sách các số epochs tương ứng với độ dài của loss
epochs = range(len(loss))

```

```python
x_for_predictions = np.linspace(0,50,1000)
```

```python
y_predicted = model.predict(x_for_predictions)
```

```python
plt.plot(x, y, '*')

plt.plot(x_for_predictions, y_predicted, 'r')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Dự đoán giá trị y dựa trên x
prediction_for_true_y = model.predict(x)
```

```python
# Tính Mean Squared Error (MSE)
mse = mean_squared_error(y, prediction_for_true_y)
print("Mean Squared Error (MSE):", mse)
```

```python
# Tính Mean Absolute Error (MAE)
mae = mean_absolute_error(y, prediction_for_true_y)
print("Mean Absolute Error (MAE):", mae)
```
